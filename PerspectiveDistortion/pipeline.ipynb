```python
for file_num, filename in enumerate(os.listdir(data_path)):
    print()
    print("Processing file #", file_num)
    full_file_path = os.path.join(data_path, filename)
    if not os.path.isfile(full_file_path):
        continue
    print(f"file: {full_file_path}")
    
    # perspective
    perspective_pipe.process_file(full_file_path, verbose=verbose)
    
    # 
    
    process = subprocess.run(["mkdir", data_path + f"example{file_num}"])
    subprocess.call(f"mv *.jpg {data_path}example{file_num}", shell=True)
#     if file_num == 0:
#         break
```
